### Data cleaning and Preparation ###

```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

```python
(df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
```

```python
df[df['Customer Lname'].isnull()][['Customer Id',
                                   'Customer Fname',
                                   'Customer Lname']].head(10)
```

```python
df['Customer Lname'] = df['Customer Lname'].fillna('Missing')
```

```python
df['Customer Lname'].isnull().sum()
```

```python
df[df['Customer Zipcode'].isnull()][['Customer Id',
                                     'Customer City',
                                     'Customer State',
                                     'Customer Zipcode']]
```

```python
df['Customer Zipcode'] = df['Customer Zipcode'].fillna(0)
```

```python
df['Customer Zipcode'].isnull().sum()
```

```python
df[['Product Description', 'Order Zipcode']].head()
```

```python
df = df.drop(columns=['Product Description', 'Order Zipcode'])
```

```python
df.shape
```

```python
df[['order date (DateOrders)',
    'shipping date (DateOrders)']].dtypes
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
```

```python
df[['order date (DateOrders)',
    'shipping date (DateOrders)']].dtypes
```

```python
df['order date (DateOrders)'].dt.year
```

```python
df['order date (DateOrders)'].dt.month
```

```python
df['order date (DateOrders)'].dt.quarter
```

```python
df['Order Year'] = df['order date (DateOrders)'].dt.year

df['Order Month'] = df['order date (DateOrders)'].dt.month

df['Order Quarter'] = df['order date (DateOrders)'].dt.quarter
```

```python
df[['order date (DateOrders)',
    'Order Year',
    'Order Month',
    'Order Quarter']].head()
```

```python
df['Shipping Delay'] = (
    df['Days for shipping (real)']
    - df['Days for shipment (scheduled)']
)
```

```python
df[['Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Shipping Delay']].head()
```

```python
df['On Time Delivery'] = df['Shipping Delay'].apply(
    lambda x: 'Yes' if x <= 0 else 'No'
)
```

```python
df[['Shipping Delay', 'On Time Delivery']].head()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

### Feature 1: Profit Margin %

```python
df['Profit Margin %'] = (
    df['Order Profit Per Order']
    / df['Sales']
) * 100
```

```python
df[['Sales',
    'Order Profit Per Order',
    'Profit Margin %']].head()
```

### Feature 2: Delivery Performance

```python
df['Delivery Performance'] = df['Shipping Delay'].apply(
    lambda x: 'Early' if x < 0 else ('On Time' if x == 0 else 'Late')
)
```

```python
df[['Shipping Delay',
    'Delivery Performance']].head()
```

```python
import matplotlib.pyplot as plt

top_regions = df.groupby('Order Region')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_regions.plot(kind='barh')
plt.title('Top 10 Regions by Sales')
plt.xlabel('Sales')
plt.ylabel('Order Region')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
segment_sales = df.groupby('Customer Segment')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
segment_sales.plot(kind='bar')

plt.title('Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
delivery_counts = df['Delivery Performance'].value_counts()

plt.figure(figsize=(8,5))
delivery_counts.plot(kind='bar')

plt.title('Delivery Performance Distribution')
plt.xlabel('Delivery Performance')
plt.ylabel('Number of Orders')

plt.tight_layout()
plt.show()
```
